# **import packages** 

```python
import cv2 
import numpy as np 
import handtrackingmodule as htm 
import math 
```

```python
from pynput.keyboard import Key,Controller
keyboard = Controller()
```

# **call object from handDetector class in handtrackingmodule module**

```python
detector = htm.handDetector(detectioncon=0.7)
```

# **Define some variables to use it Later**

**to get minAngle and maxAngel values that fits your device , please run the code in "get_range" file** 

```python
pTime = 0
last_angle = None 
last_length = None 
minAngle = -65.25
maxAngle = 0 
angle = 0 
angleBar = 400 
angleDeg = 0 
minHand = 50
maxHand = 300
```

# **here we go , the main code :** 

```python
cap = cv2.VideoCapture(0)  # Start capturing video from the default camera (index 0)

while True: 
    ret, img = cap.read()  # Read a frame from the video feed
    img = detector.findHands(img)  # Detect and draw hands on the frame using a hand detection model
    lmList = detector.findPositions(img, draw=False)  # Get the list of landmark positions for the detected hand(s)
    
    # Uncomment the following lines to print the positions of thumb (index 4) and index finger (index 8)
    # if len(lmList) != 0:  
    #     print(lmList[4], lmList[8])

    if len(lmList) > 8:  # Ensure at least the thumb (4th landmark) and index finger (8th landmark) are detected
        x1, y1 = lmList[4][0], lmList[4][1]  # Get the coordinates of the thumb tip
        x2, y2 = lmList[8][0], lmList[8][1]  # Get the coordinates of the index finger tip
        cx, cy = (x1+x2)//2, (y1+y2)//2  # Calculate the center point between the thumb and index finger
        
        # Draw red circles on the thumb and index finger tips
        cv2.circle(img, (x1, y1), 15, (0, 0, 255), cv2.FILLED)
        cv2.circle(img, (x2, y2), 15, (0, 0, 255), cv2.FILLED)
        
        # Draw a green line connecting the thumb and index finger
        cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 3)
        
        # Calculate the distance (length) between the thumb and index finger tips
        length = math.hypot(x2 - x1, y2 - y1)
        
        # Map the length of the hand gesture to the desired volume control range
        angle = np.interp(length, [minHand, maxHand], [minAngle, maxAngle])
        angleBar = np.interp(length, [minHand, maxHand], [400, 150])
        angleDeg = np.interp(length, [minHand, maxHand], [0, 100])
        angleDeg = round(angleDeg)  # Round the angle value to an integer for display
        
        # Check if the last frame has a recorded distance (last_length) for comparison
        if last_length: 
            # If the new length is greater than the previous length, increase the volume
            if length > last_length: 
                keyboard.press(Key.media_volume_up)
                keyboard.release(Key.media_volume_up)
                print("Volume Up")
            # If the new length is smaller, decrease the volume
            elif length < last_length: 
                keyboard.press(Key.media_volume_down)
                keyboard.release(Key.media_volume_down)
                print("Volume Down")
        
        # Update the last measured angle and length for comparison in the next frame
        last_angle = angle 
        last_length = length
        
        # Visual feedback: If the length is small (gesture closed), make the center circle red, otherwise green
        if length <= 50: 
            cv2.circle(img, (cx, cy), 15, (0, 0, 255), cv2.FILLED)
        elif length > 50: 
            cv2.circle(img, (cx, cy), 15, (0, 255, 0), cv2.FILLED)
        
        # Draw a volume control bar on the side of the screen
        cv2.rectangle(img, (50, 150), (85, 400), (255, 0, 0), 3)  # Outer bar (static)
        cv2.rectangle(img, (50, int(angleBar)), (85, 400), (255, 0, 0), cv2.FILLED)  # Inner bar representing the volume level
        
        # Display the percentage of volume (angleDeg) as text on the screen
        cv2.putText(img, str(angleDeg) + "%", (40, 120), cv2.FONT_HERSHEY_PLAIN, 2, (255, 0, 0), 3)

    # Display the processed video frame in a window
    cv2.imshow("img", img)
    
    # Break the loop and stop the video feed if the ESC key (27) is pressed
    if cv2.waitKey(1) == 27:
        break

cap.release()  # Release the camera resource
out.release()  # Release any other output resources if used (if saving video, for example)
cv2.destroyAllWindows()  # Close all OpenCV windows
```
