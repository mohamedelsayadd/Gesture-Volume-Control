```python
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
from ctypes import cast, POINTER

# Get the default audio playback device (speakers)
devices = AudioUtilities.GetSpeakers()

# Activate the IAudioEndpointVolume interface on the device with CLSCTX_ALL
interface = devices.Activate(IAudioEndpointVolume._iid_, CLSCTX_ALL, None)

# Cast the interface to access IAudioEndpointVolume methods
volume = cast(interface, POINTER(IAudioEndpointVolume))

# Get and print the volume range (min, max, step)
print(volume.GetVolumeRange())
```
